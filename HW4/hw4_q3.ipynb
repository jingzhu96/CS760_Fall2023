```python
import pandas as pd
import numpy as np
import math

characters =['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z',' ']
language = ['e','j','s']

def read_file(filename, path = './languageID/'):
    file = open(path+filename, 'r')
    file_content = file.read()
    file_content = file_content.replace('\n', '')
    return file_content

def index(c):
    if c == ' ':
        return 26
    else:
        return ord(c)-ord('a')

def count_character(file):
    c_count = []
    l = len(file)
    for i in characters:
        count = 0
        for j in range(l):
            for k in file[j]:
                if i == k:
                    count+=1
        c_count.append(count)
    return c_count
```

```python
e_list=[]
j_list=[]
s_list=[]
for i in range(10):
    e_list.append(read_file("e"+str(i)+'.txt'))
    j_list.append(read_file("j"+str(i)+'.txt'))
    s_list.append(read_file("s"+str(i)+'.txt'))
```

### (2)

```python
# Q2
df = pd.DataFrame(characters, columns=['characters'])
# theta_e
df['e_count'] = count_character(e_list)
total_e = sum(count_character(e_list))

theta_log_e = [math.log(count_character(e_list)[i]+1/2) - math.log(total_e+27/2) for i in range(27)]
theta_e = [math.e ** theta_log_e[i] for i in range(27)]
df['e_theta'] = theta_e
df['e_theta_log'] = theta_log_e
```

```python
# Q3
# theta_j
df['j_count'] = count_character(j_list)
total_j = sum(count_character(j_list))
theta_log_j = [math.log(count_character(j_list)[i]+1/2) - math.log(total_j+27/2) for i in range(27)]
theta_j = [math.e ** theta_log_j[i] for i in range(27)]
df['j_theta'] = theta_j
df['j_theta_log'] = theta_log_j

# theta_s
df['s_count'] = count_character(s_list)
total_s = sum(count_character(s_list))
theta_log_s = [math.log(count_character(s_list)[i]+1/2) - math.log(total_s+27/2) for i in range(27)]
theta_s = [math.e ** theta_log_s[i] for i in range(27)]
df['s_theta'] = theta_s
df['s_theta_log'] = theta_log_s
```

### (7)

```python
def predict(count_list):
    theta = []
    pred = []
    for count in count_list:
        p_x_y=[]
        for l in language:
            col = l+'_theta_log'
            theta_log = [(count[i] * df[col][i]) for i in range(27)]
            p_x_y.append(sum(theta_log))
        theta.append(p_x_y)
    
    for t in theta:
        if t[0]>=t[1] and t[0]>=t[2]:
            p = 'e'
        elif t[1]>=t[0] and t[1]>=t[2]:
            p = 'j'
        elif t[2]>=t[0] and t[2]>=t[1]:
            p = 's'
        pred.append(p)
    return pred
```

```python
e_count_list = []
j_count_list = []
s_count_list = []
for i in range(10,20):
    e_count_list.append(count_character(read_file("e"+str(i)+'.txt')))
    j_count_list.append(count_character(read_file("j"+str(i)+'.txt')))
    s_count_list.append(count_character(read_file("s"+str(i)+'.txt')))
count_list = [e_count_list, j_count_list, s_count_list]

predicted = []
predicted_labels = []
for i in range(3):
    predicted_label = predict(count_list[i])
    predicted.append(predicted_label)
    predicted_labels.append([language.index(e) for e in predicted_label])
predicted
```

```python
gt = [['e']*10, ['j']*10, ['s']*10]
ground_truth = []
for i in range(3):
   ground_truth.append([language.index(e) for e in gt[i]])
```

```python
confusion_matrix = np.zeros((3,3))
for i in range(3):
    for j in range(len(predicted_labels[i])):
        p = predicted_labels[i][j]
        t = ground_truth[i][j]
        confusion_matrix[p][t] += 1

print("Confusion Matrix:\n", confusion_matrix)
```
