```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
D1 = pd.read_csv('D1.txt', sep=' ', header=None)
D2 = pd.read_csv('D2.txt', sep=' ', header=None)
D1.columns = ['x1', 'x2', 'y']
D2.columns = ['x1', 'x2', 'y']
```

```python
fig1 = plt.figure(figsize=(8, 8))
colors = {1: 'red', 0: 'blue'}
sizes = {1: 25, 0: 25} 

# Map colors to the 'y' column
D1['color'] = D1['y'].map(colors)

# Create the scatter plot
plt.scatter(D1['x1'], D1['x2'], c=D1['color'], s=D1['y'].map(sizes), label=D1['y'])

# Add labels and legend
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', label='Label 1', markerfacecolor='red', markersize=8),
                    plt.Line2D([0], [0], marker='o', color='w', label='Label 0', markerfacecolor='blue', markersize=8)])

# Show the plot
plt.show()
```

```python
fig2 = plt.figure(figsize=(8, 8))
colors = {1: 'red', 0: 'blue'}
sizes = {1: 25, 0: 25} 

# Map colors to the 'y' column
D2['color'] = D2['y'].map(colors)

# Create the scatter plot
plt.scatter(D2['x1'], D2['x2'], c=D2['color'], s=D2['y'].map(sizes), label=D2['y'])

# Add labels and legend
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', label='Label 1', markerfacecolor='red', markersize=8),
                    plt.Line2D([0], [0], marker='o', color='w', label='Label 0', markerfacecolor='blue', markersize=8)])

# Show the plot
plt.show()
```
