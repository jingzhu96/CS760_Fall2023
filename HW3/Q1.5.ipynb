```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
confidences = [0.95, 0.85, 0.8, 0.7, 0.55, 0.45, 0.4, 0.3, 0.2, 0.1]
classes = [1, 1, 0, 1, 1, 0, 1, 1, 0, 0]

tpr_list = []
fpr_list = []
positive_count = classes.count(1)
negative_count = classes.count(0)

for threshold in confidences:
    tp = 0
    fp = 0
    tn = 0
    fn = 0
    for i, confidence in enumerate(confidences):
        if confidence >= threshold:
            if classes[i] == 1:
                tp += 1
            else:
                fp += 1
        else:
            if classes[i] == 1:
                fn += 1
            else:
                tn += 1
    
    tpr = tp / positive_count
    fpr = fp / negative_count
    
    tpr_list.append(tpr)
    fpr_list.append(fpr)
```

```python
fig0 = plt.figure(figsize=(8, 6))
plt.plot(fpr_list, tpr_list, marker='o', color='b')
plt.plot([0, 1], [0, 1], linestyle='--', color='r')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.grid(True)
plt.show()
fig0.savefig('P1.5_ROC_curve.png')
```
